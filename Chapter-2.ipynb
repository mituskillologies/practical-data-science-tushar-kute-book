```python
import numpy as np

arr2d = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])
print("Original 2D array:\n", arr2d)

# दुसऱ्या ओळीतील तिसरा घटक (इंडेक्स 1, 2)
print("Element at row 1, column 2:", arr2d[1, 2])

# फक्त दुसरी ओळ
print("Second row:", arr2d[1, :]) # : चा ॲरे्थ 'सर्व स्तंभ'

# फक्त तिसरा स्तंभ
print("Third column:\n", arr2d[:, 2]) # : चा ॲरे्थ 'सर्व ओळी'

# उप-ॲरेे (Sub-array)
# पहिल्या दोन ओळी आणि दुसरे-तिसरे स्तंभ
sub_arr = arr2d[:2, 1:]
print("Sub-array (first 2 rows, columns 1 and 2):\n", sub_arr)
```

```python
import pandas as pd
from datetime import datetime

# आपण आधी वापरलेलाच कंपन्यांचा डेटाफ्रेम घेऊया
company_data = {
    'company_name': ['Reliance', 'TCS', 'HDFC Bank', 'Infosys'],
    'industry': ['Conglomerate', 'IT', 'Banking', 'IT'],
    'foundation_year': [1973, 1968, 1994, 1981]
}
df = pd.DataFrame(company_data)

# सध्याचे वर्ष मिळवणे
current_year = datetime.now().year

# 'company_age' नावाचा नवीन स्तंभ तयार करणे
df['company_age'] = current_year - df['foundation_year']

print("नवीन स्तंभ जोडल्यानंतरचा डेटाफ्रेम:\n")
print(df)
```

```python
import pandas as pd
import numpy as np # NaN तयार करण्यासाठी NumPy ची मदत घेऊ

exam_scores = {
    'student_name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Vivek'],
    'math_score': [85, 90, np.nan, 75, 92],
    'science_score': [91, 88, 82, 94, np.nan],
    'history_score': [80, 78, 88, np.nan, 85]
}
scores_df = pd.DataFrame(exam_scores)
print("गहाळ माहिती असलेला मूळ डेटाफ्रेम:\n")
print(scores_df)
```

```python
# दोन लहान डेटाफ्रेम तयार करूया
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']})
	
# त्यांना एकाखाली एक जोडणे (vertical stacking)
vertical_concat = pd.concat([df1, df2])
print("एकाखाली एक जोडलेले डेटाफ्रेम (Vertical Concat):\n")
print(vertical_concat)

# इंडेक्स रिसेट करणे
vertical_concat_reset = pd.concat([df1, df2], ignore_index=True)
print("\nइंडेक्स रिसेट करून जोडलेले डेटाफ्रेम:\n")
print(vertical_concat_reset)
```

```python
# Outer Join
outer_join_df = pd.merge(students, scores, on='student_id', how='outer')
print("\nOuter Join:\n")
print(outer_join_df)
```

```python
# Left Join
left_join_df = pd.merge(students, scores, on='student_id', how='left')
print("\nLeft Join:\n")
print(left_join_df)
```

```python
import pandas as pd
import numpy as np

# Anscombe's Quartet डेटासेट तयार करणे
anscombe = pd.DataFrame({
    'x1': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y1': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68],
    'x2': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y2': [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74],
    'x3': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y3': [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73],
    'x4': [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8],
    'y4': [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
})

# प्रत्येक गटासाठी सांख्यिकीय माहिती पाहूया
# पहिला गट
dataset1 = anscombe[['x1', 'y1']]
# दुसरा गट
dataset2 = anscombe[['x2', 'y2']]
dataset2.columns = ['x', 'y']
# तिसरा गट
dataset3 = anscombe[['x3', 'y3']]
dataset3.columns = ['x', 'y']
# चौथा गट
dataset4 = anscombe[['x4', 'y4']]
dataset4.columns = ['x', 'y']


print("Dataset 1 Statistics:\n", dataset1.describe())
print("\nDataset 2 Statistics:\n", dataset2.describe())
print("\nDataset 3 Statistics:\n", dataset3.describe())
print("\nDataset 4 Statistics:\n", dataset4.describe())
```

```python
import matplotlib.pyplot as plt

# एक फिगर आणि 2x2 सबप्लॉट्स (axes) चा ग्रिड तयार करणे
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# प्रत्येक सबप्लॉटला एक शीर्षक देणे
fig.suptitle("Anscombe's Quartet - The Power of Visualization", fontsize=16)

# पहिला डेटासेट प्लॉट करणे
axes[0, 0].scatter(anscombe['x1'], anscombe['y1'])
axes[0, 0].set_title("Dataset 1")

# दुसरा डेटासेट प्लॉट करणे
axes[0, 1].scatter(anscombe['x2'], anscombe['y2'])
axes[0, 1].set_title("Dataset 2")

# तिसरा डेटासेट प्लॉट करणे
axes[1, 0].scatter(anscombe['x3'], anscombe['y3'])
axes[1, 0].set_title("Dataset 3")

# चौथा डेटासेट प्लॉट करणे
axes[1, 1].scatter(anscombe['x4'], anscombe['y4'])
axes[1, 1].set_title("Dataset 4")

# लेआउट व्यवस्थित करणे
plt.tight_layout()
# शीर्षक आणि प्लॉटमध्ये योग्य अंतर राखणे
fig.subplots_adjust(top=0.9)

# आलेख दाखवणे
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 100 विद्यार्थ्यांचे काल्पनिक गुण (0 ते 100)
student_scores = np.random.randint(0, 101, 100)

fig, ax = plt.subplots()
ax.hist(student_scores, bins=10, edgecolor='black') # bins म्हणजे किती गटांमध्ये विभागणी करायची

ax.set_title("Distribution of Student Scores")
ax.set_xlabel("Score")
ax.set_ylabel("Number of Students")

plt.show()
```

```python
# डेटा
fruits = ['Mango', 'Apple', 'Banana']
students = [70, 30, 40]

# पाय चार्ट तयार करणे
plt.pie(students, labels=fruits, autopct='%1.1f%%', startangle=90)

# शीर्षक जोडणे
plt.title("Favorite Fruits of Students")

# समान वर्तुळाकार आकार
plt.axis('equal')  

# चार्ट दाखवणे
plt.show()
```

```python
# डेटा तयार करणे
# Class A चे गुण (आकडे थोडे कमी-जास्त आहेत)
class_a_scores = [75, 80, 85, 90, 92, 95, 98, 70, 65, 88]

# Class B चे गुण (आकडे थोडे जास्त पसरलेले आहेत)
class_b_scores = [60, 65, 70, 75, 80, 85, 90, 95, 100, 50, 45, 105, 110]

# दोन्ही डेटा एका लिस्टमध्ये एकत्र करणे
data_to_plot = [class_a_scores, class_b_scores]

# बॉक्स प्लॉट तयार करणे
fig, ax = plt.subplots()
ax.boxplot(data_to_plot, labels=['Class A', 'Class B'])

# शीर्षक आणि लेबल्स जोडणे
ax.set_title('Comparison of Math Scores between Two Classes')
ax.set_xlabel('Classes')
ax.set_ylabel('Scores')

# चार्ट दाखवणे
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# काही कंपन्यांचा डेटा
df_companies = pd.DataFrame({
    'industry': ['IT', 'IT', 'IT', 'Banking', 'Banking', 'Pharma', 'Pharma'],
    'employee_satisfaction': [8.2, 7.8, 8.5, 7.1, 7.5, 9.0, 8.8]
})

fig, ax = plt.subplots()
sns.boxplot(x='industry', y='employee_satisfaction', data=df_companies, ax=ax)
ax.set_title("Employee Satisfaction by Industry")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Iris डेटासेट, जो फुलांच्या पाकळ्यांची माहिती देतो, तो Seaborn मध्ये उपलब्ध असतो
iris = sns.load_dataset("iris")

# species नुसार रंगांची विभागणी करणे
sns.pairplot(iris, hue="species")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# एक काल्पनिक कोरिलेशन मॅट्रिक्स तयार करणे
data = pd.DataFrame(np.random.rand(5, 5), columns=['A', 'B', 'C', 'D', 'E'])
correlation_matrix = data.corr()

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', ax=ax) # annot=True म्हणजे प्रत्येक सेलमध्ये व्हॅल्यू दाखवणे
ax.set_title("Correlation Heatmap")
plt.show()
```
